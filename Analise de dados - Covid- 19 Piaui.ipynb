```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data_path ='/Users/JOSE MARIA/Downloads/6° periodo/Sistemas inteligentes'
data_name ='caso.csv'
```

```python
def load_csv(path,name):
    arquivo = os.path.join(path,name)
    return pd.read_csv(arquivo)
data = load_csv(data_path,data_name)
```

```python
#1
filtro = data.loc[(data.is_last==True) & (data.place_type == 'city') & (data.state=='PI')]
```

```python
cidades= pd.read_csv('municipios.csv')
```

```python
cidades
```

```python
cidades = cidades[['longitude','latitude','nome']].loc[cidades.codigo_uf==22]
#22 é o codigo uf do Piauí
```

```python
cidades = cidades.rename({'nome':'city'},axis=1)#renomeando atributo para a junção interna
```

```python
cidades
```

```python
#5
filtro = pd.merge(filtro,cidades,how='inner',on='city')#fazendo junção interna de colunas
```

```python
filtro.head(5)
```

```python
#6
filtro.plot(kind='scatter',x='latitude',y='longitude',sharex=False,c='death_rate',label='Population',figsize=(10,7),
            s=filtro['estimated_population']/500,colorbar=True,cmap=plt.get_cmap('jet'),alpha=0.3)
plt.legend()
plt.show()
```
